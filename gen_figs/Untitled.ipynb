```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'

key_sizes = [128, 256, 512, 1024, 2048, 3072]

x = np.arange(0,len(key_sizes)*2,2)
width=0.5
x1 = x-width/2
x2 = x+width/2
y1 = [0.000009, 0.000029, 0.000179, 0.001201, 0.008316, 0.026309]

y2 = [0.000007, 0.000015, 0.000058, 0.000380, 0.002594, 0.007753]

y1 = [1000* x for x in y1]
y2 = [1000* x for x in y2]
# 绘制分组柱状图
plt.bar(x1,y1,width=0.5,label='phe',color='#f9766e',edgecolor='k',zorder=2)
plt.bar(x2,y2,width=0.5,label='iphe',color='#00bfc4',edgecolor='k',zorder=2)

# 添加x,y轴名称、图例和网格线
plt.xlabel('dec time',fontsize=11)
plt.ylabel('Time(s)',fontsize=11)
plt.legend(frameon=False)
plt.grid(ls='--',alpha=0.8)

# 修改x刻度标签为对应日期
plt.xticks(x, key_sizes ,fontsize=10)
plt.tick_params(axis='x',length=0)


plt.tight_layout()
plt.savefig('crt_phe.pdf',dpi=600)
plt.show()
```
