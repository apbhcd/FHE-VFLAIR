```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_time_arr(tarr):
    res = []
    for i in range(0, len(tarr)):
        res.append(tarr[i] - tarr[0])
    return res
```

```python
x = np.linspace(0, epochs, epochs)
```

```python
# plt.title('Result Analysis')
plt.plot(x, plain_breast_cancer_time, 'b^-', label='Vanilla-FedAvg Train Accuracy')
plt.plot(x, paillier_3072_breast_cancer_time, 'ro-',label='Paillier-FedAvg Train Accuracy')
plt.plot(x, crt_paillier_3072_breast_cancer_time, 'g+-', label='BatchAgg Train Accuracy')
plt.legend() # 显示图例
plt.xlabel('Epochs')
plt.ylabel('Train Accuracy on CIFAR-10')
plt.grid()
plt.rcParams['savefig.dpi'] = 600 #图片像素
plt.rcParams['figure.dpi'] = 600 #分辨率
plt.savefig('secureboost_time.png', dpi=600)
plt.savefig('secureboost_time.pdf',format="pdf", dpi=600)
plt.show()
```

```python

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
width = 0.2
labels = [128, 256, 512, 1024, 2048, 4096, 8192] # 级别
men_means = np.random.randint(20,35,size = 7)
women_means = np.random.randint(20,35,size = 7)

x = np.arange(len(men_means))

plt.figure(figsize=(9,7))

rects1 = plt.bar(x - width/2, men_means, width) # 返回绘图区域对象
rects2 = plt.bar(x + width/2, women_means, width)

# 设置标签标题，图例
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(x,labels)
plt.legend(['Men','Women'])

# 添加注释
def set_label(rects):
    for rect in rects:
        height = rect.get_height() # 获取⾼度
        plt.text(x = rect.get_x() + rect.get_width()/2, # ⽔平坐标
                 y = height + 0.5, # 竖直坐标
                 s = height, # ⽂本
                 ha = 'center') # ⽔平居中

set_label(rects1)
set_label(rects2)

plt.tight_layout() # 设置紧凑布局
plt.savefig('./分组带标签柱状图.png')
```
